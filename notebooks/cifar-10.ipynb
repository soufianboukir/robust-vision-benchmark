# Dataset used - CIFAR-10

The **CIFAR-10** dataset is a widely used benchmark in machine learning and computer vision, consisting of **60,000 32x32 color** images across 10 distinct classes (e.g., airplanes, cars, cats, dogs). 

It includes 50,000 training images and 10,000 test images, with **6,000 images per class**. Developed by **Alex Krizhevsky**, **Vinod Nair**, and **Geoffrey Hinton**, it is a staple for training image classification models. 

- Structure: 10 classes: airplane, automobile, bird, cat, deer, dog, frog, horse, ship, truck.
- Image Dimensions: 32x32 color images (pixels).
- Dataset Split: 50,000 training images, 10,000 testing images.
- Usage: Frequently used in CNN (Convolutional Neural Network) tasks to evaluate image classification performance.

```python
import torch
import matplotlib.pyplot as plt
import numpy as np
from torchvision import datasets, transforms
import random



transform = transforms.ToTensor()

train_dataset = datasets.CIFAR10(
    root="../data/raw",
    train=True,
    download=True,
    transform=transform
)

classes = [
    "airplane", "automobile", "bird", "cat", "deer",
    "dog", "frog", "horse", "ship", "truck"
]



def show_random_samples(dataset, num_images=12):
    plt.figure(figsize=(10, 6))

    indices = random.sample(range(len(dataset)), num_images)

    for i, idx in enumerate(indices):
        image, label = dataset[idx]

        # Convert tensor (C,H,W) → (H,W,C)
        image = image.permute(1, 2, 0).numpy()

        plt.subplot(3, 4, i + 1)
        plt.imshow(image)
        plt.title(classes[label])
        plt.axis("off")

    plt.tight_layout()
    plt.savefig("../results/imgs/cifar-10.png", dpi=150, bbox_inches="tight")
    plt.show()


show_random_samples(train_dataset, num_images=12)
```
